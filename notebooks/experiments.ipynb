```python
import re

def count_personal_pronouns(text):
    # Define the list of pronouns and the pattern
    pronouns = ["I", "we", "my", "ours", "us"]
    pattern = r'\b(?:I|we|my|ours|us)\b'
    
    # Compile the regular expression pattern with case-insensitive matching
    regex = re.compile(pattern, re.IGNORECASE)
    
    # Find all matches in the text
    matches = regex.findall(text)
    
    # Count occurrences of each pronoun
    pronoun_counts = {pronoun: 0 for pronoun in pronouns}
    
    for match in matches:
        # Normalize to lowercase for counting purposes
        match_lower = match.lower()
        if match_lower in pronoun_counts:
            pronoun_counts[match_lower] += 1

    return pronoun_counts

# Example usage
text = "I went to the US. We visited the museum. My friend and I talked about our plans. The US is a great country. Us being here makes it better."

pronoun_counts = count_personal_pronouns(der)

print("Personal Pronoun Counts:")
for pronoun, count in pronoun_counts.items():
    print(f"'{pronoun}': {count}")
```
